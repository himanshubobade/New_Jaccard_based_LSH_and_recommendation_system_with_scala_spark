```python
import pandas as pd
import numpy as np

# Load the first CSV file into a DataFrame
df1 = pd.read_csv('himanshu_task1.csv', header=0)

# Add a column 'csv1' to df1
#df1['csv1'] = 'csv1'

# Load the second CSV file into a DataFrame
df2 = pd.read_csv('task1_kanchi.csv', header=0)

# Add a column 'csv2' to df2
#df2['csv2'] = 'csv2'

# Concatenate the two DataFrames
df_concat = pd.concat([df1, df2])

df1_sorted = df1.sort_values(by=['business_id_1',' business_id_2'])
df2_sorted = df2.sort_values(by=['business_id_1',' business_id_2'])
```

```python
df1_sorted.to_csv('df1_sorted.csv', index=False)
df2_sorted.to_csv('df2_sorted.csv', index=False)
```
